```python
import cv2
import os
import time
import uuid
```

```python
IMAGES_PATH = 'E:\Projects\Real time sign language detection\RealTimeObjectDetection-main\RealTimeObjectDetection-main\Tensorflow\workspace\images\collectedimages'
```

```python
labels = ['hello', 'thanks', 'yes', 'no', 'iloveyou']
number_imgs = 15
```

```python
for label in labels:
    label_path = os.path.join(IMAGES_PATH, label)
    os.makedirs(label_path, exist_ok=True)

    cap = cv2.VideoCapture(0)

    print(f'Collecting images for {label}')
    time.sleep(5)

    for img_num in range(number_imgs):
        ret, frame = cap.read()
        if not ret:
            print("Failed to grab frame")
            break

        imgname = os.path.join(label_path, f'{label}.{str(uuid.uuid1())}.jpg')
        cv2.imwrite(imgname, frame)
        cv2.imshow('frame', frame)
        time.sleep(2)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
```
